```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy import stats
from statsmodels.stats.power import TTestIndPower
from tqdm import tqdm
```

# Описание


Основной фактор выбора критерия для проведения АБ-теста, при заданной метрики, является минимальный размер выборки при заданных $\alpha$ и $1-\beta$. 

Например: при $\alpha = 0.05$ и $1-\beta = 0.8$ для критерия А нужна выборка размера 100к, для критерия Б выборка 150к. Критерий А предпочтительней потому, что АБ-тест по времени будет длится меньше.

## 1. Количественные метрики

Количественная метрика - это числовая мера, которая используется для измерения и количественной оценки различных аспектов продукта или пользовательского опыта.

Примеры количественных метрик включают величина среднего чека, время прохождения уровня, средняя продолжительность сессии, количество опыта и так далее.

Посмотрим на t-test и U-критерий Манна-Уитни, посчитаем размер выборки, посмотрим на показатель $\alpha$ при АА-тесте и $1-\beta$ при АБ-тесте. 

### 1.1 t-test Стьюдента

- чтобы выборочные средние имели нормальное распределение. При маленьких выборках это означает требование нормальности исходных значений. 
- В случае применения двухвыборочного критерия для независимых выборок также необходимо соблюдение условия равенства дисперсий. Существуют, однако, альтернативы критерию Стьюдента для ситуации с неравными дисперсиями.
- Также не вполне корректно применять t-критерий Стьюдента при наличии в данных значительного числа выбросов

**1.1.1** Проверим первое ограничение t-теста, а именно "нормальное распределение выборочных средних"

```python
# напишем функцию генерации данных 
# из Пуассоновского распределения

def generate_data(mu1, mde, sample_size, func, range_size=1000):
    result = []
    for _ in tqdm(range(range_size)):
        a = np.random.poisson(lam=mu1, size=sample_size)
        b = np.random.poisson(lam=mu1+mde, size=sample_size)
        
        result.append(func(a, b))
    
    return result

mu1 = 2
mde = 0.3

func = lambda x, y: (np.mean(x), np.mean(y))

means_array = generate_data(mu1=mu1, 
                            mde=mu1+mde,
                            sample_size=1000,
                            func=func)
```

```python
# построим гистограмму данных и граффик qqplot,
# который показывает связь теоретических квантилей
# нормального распределения и квантилей нашей выборки

means_array = np.array(means_array)

fig, axs = plt.subplots(2, 2, figsize=(10,8))

bins = 30
axs[0, 0].hist(means_array[:, 0], bins=bins)
axs[0, 1].hist(means_array[:, 1], bins=bins)

stats.probplot(means_array[:, 0], dist='norm', plot=axs[1,0])
stats.probplot(means_array[:, 1], dist='norm', plot=axs[1,1])


axs[0,0].set_title('hist a')
axs[0,1].set_title('hist b')

axs[1,0].set_title('qqplot a')
axs[1,1].set_title('qqplot b')

plt.show();


# и проверим гипотезу о нормальности распределения для каждой выборки

# Сформируем нулевую и альтернативную гипотезу:
# H0: выборка из нормального распределения
# H1: выборка не из нормального распределения

alpha = 0.05
_, pval1 = stats.shapiro(means_array[:, 0])
_, pval2 = stats.shapiro(means_array[:, 1])

print(f'Выборка 1: p_value = {pval1:0.3f}, H0 {"не отвергаем" if pval1 > alpha else "отвергаем"}\n'
     f'Выборка 2: p_value = {pval2:0.3f}, H0 {"не отвергаем" if pval2 > alpha else "отвергаем"}')
```

Граффик гистограм показывает колокообразное, симетричное распределения с небольшими выбросами в хвостах. Граффик qqplot показывает как эмпирические квантили идеально совпадают с теоретическими, единственно в хвостах есть немного выбросов. И наконец, критерий согласия Шипиро-Уилко, который проверяет нормальность распределения, дает p-value в районе 1 для обеих выборок, что не дает нам права отвергнуть нулевую гипотезу.

**1.1.2** Второе требование говорит о равентсве дисперсии двух выборок. При разной дисперсии, нужно использовать t-test Уэлча. Отличаются они формулой нахождения степеней свободы для t распределения. При достаточно больших выборках, t статистика Стьюдента и t статистика Уэлча стремиться к нормальному распределению. 

Давайте посмотрим как ведет себя p-value при разных дисперсиях на разных обьемах выборок при использовании теста Стьюдента, Уэлча и теста Стюдента в котором для подсчета p-value будем брать квантили нормального распределения.

```python
def t_test(a_sample, b_sample):
    a_m = np.mean(a_sample)
    b_m = np.mean(b_sample)
    a_var = np.var(a_sample)
    b_var = np.var(b_sample)
    a_size= a_sample.size
    b_size = b_sample.size
    
    t_stat = (a_m - b_m) / (a_var/a_size + b_var/b_size)**(1/2) # считаем t статистику
    p_value = (1 - stats.norm().cdf(abs(t_stat))) * 2 # считаем p-value:
                                                      # (1 - модуль t стат.) считаем плотность веростности от
                                                      # t стат. до бесконечности
                                                      # умножаем на 2 при проверки двухсторонней гипотезы
    
    return t_stat, p_value
```

```python
# cнегерируем выборки из распределения Пуассона с 
# разными параметрами lambda, которые являются и
# мат ожиданием и дисперсией распределения.

mu1 = 2
mde = 0.5

p_vals_st = []
p_vals_welch = []
p_vals_norm = []
for i in tqdm(range(10, 1000, 10)):
    a = np.random.poisson(lam=mu1, size=i)
    b = np.random.poisson(lam=mu1+mde, size=i)
    
    _, pval_st = stats.ttest_ind(a, b, equal_var=True)     # t-test для одинаковой дисперсии
    _, pval_welch = stats.ttest_ind(a, b, equal_var=False) # t-test для разной дисперсии
    _, pval_norm = t_test(a, b)                            # t-test для одинаковой дисперсии, 
                                                           # p_value посчитано на нормальном распределении
    
    p_vals_st.append(pval_st)
    p_vals_welch.append(pval_welch)
    p_vals_norm.append(pval_norm)

df = pd.DataFrame({
    'p_vals_st': p_vals_st,
    'p_vals_welch': p_vals_welch,
    'p_vals_norm': p_vals_norm
})

df.index = list(range(10, 1000, 10))

df.head()
```

```python
plt.rcParams['lines.linewidth'] = 0.5

df.plot()

plt.title('')
plt.xlabel('размер выборки')
plt.ylabel('p-value')

plt.show();
```

```python
# небольшая разница есть в начале, чтоб ее увидеть,
# давайте возьмем за базис колонку 'p_vals_st' и
# вычтем из нее 'p_vals_welch', 'p_vals_norm'

df['welch_diff'] = df['p_vals_st'] - df['p_vals_welch']
df['norm_diff'] = df['p_vals_st'] - df['p_vals_norm']

df[['welch_diff', 'norm_diff']].plot()

plt.title('')
plt.xlabel('размер выборки')
plt.ylabel('разность p-value')

plt.show();
```

Примерно с выборки обьема 300, все три теста начинают показывать одинаковый p-value на распределении Пуассона. На других распределений нужно проверять отдельно.

**1.1.3** Третье ограничение говорит о некоректности применения t-теста при наличии выбросов в выборке. Наглядно влияние этого ограничения можно показать при симуляции АА и АБ теста для подсчета $\alpha$ и $1-\beta$, что и является целью этого ноутбука.

```python
# для начала напишем функцию для расчета размера
# выборки для t-теста

def ttest_sample_size(alpha=0.05, power=0.8, mde=0.1, std1=1,
                      std2=1, equal_var=True) -> tuple:
    """
    Расчет необходимого объема выборки для двухвыборочного t-теста.
    
    Параметры:
        alpha (float): Уровень значимости (вероятность ошибки 1-го рода).
        power (float): Желаемая мощность теста (1 - вероятность ошибки 2-го рода).
        mde (float): Минимально обнаружимый эффект (разница между средними).
        std1 (float): Стандартное отклонение первой выборки.
        std2 (float): Стандартное отклонение второй выборки.
        equal_var (bool): Предполагается ли равенство дисперсий между выборками.
    
    Возвращает:
        tuple(int, int): Кортеж с вычисленными объемами выборки (округленные значения).
    """
    if equal_var:
        effect_size = mde / std1
        div = 2 * (std1 / mde)**2
    else:
        effect_size = mde / (np.sqrt((std1**2 + std2**2) / 2))
        div = (std1**2 + std2**2) / mde**2
    
    z_alpha = stats.norm().ppf(alpha/2)
    z_beta = stats.norm().ppf(1-power)
    theoretical_ss = (z_alpha + z_beta)**2 * div
    
    ttest_power = TTestIndPower()
    sample_size = ttest_power.solve_power(effect_size=effect_size,
                                         nobs1=None,
                                         alpha=alpha,
                                         power=power,
                                         ratio=1,
                                         alternative='two-sided')
    
    return int(np.ceil(theoretical_ss)), int(np.ceil(sample_size))
```

```python
# параметры аб-теста

alpha = 0.05
power = 0.8

mu1 = 1
mde = 0.025

# для распределения Пуассона дисперсия = мат ожидание
std1 = mu1
std2 = mu1 + mde

# размер одной выборки при заданых 
# параметрах аб-теста
theoretical_ss, sample_size = ttest_sample_size(alpha=alpha, power=power,
                                               mde=mde, std1=std1, std2=std2)

print(f'Sample size = {theoretical_ss, sample_size}')
```

```python
# генерация данных для АБ-теста

data = generate_data(mu1=mu1,
                     mde=mde, 
                     sample_size=sample_size, 
                     func=t_test)

data = np.array(data)
ab_df = pd.DataFrame({
    't_stat': data[:, 0],
    'p_value': data[:, 1]
})

ab_df.head()
```

```python
# генерация данных для АA-теста

data = generate_data(mu1=mu1,
                     mde=0, 
                     sample_size=sample_size, 
                     func=t_test)

data = np.array(data)
aa_df = pd.DataFrame({
    't_stat': data[:, 0],
    'p_value': data[:, 1]
})

aa_df.head()
```

```python
def p_value_hist(data, test='AA', alpha=0.05, power=0.8):
    """
    """
    hist = data['p_value'].hist(bins=50)

    ymax = max(hist.get_ybound())
    plt.vlines(x=alpha, ymin=0, ymax=ymax, color='red')
    plt.text(x=alpha+0.02, y=ymax/10, s='alpha = 0.05', rotation=90)

    plt.title(f'{test}-тест, распределение p-value')
    plt.xlabel('p-value')
    plt.ylabel('частота')

    plt.show()

    print(f'Нормированое количество (p-value < {alpha}) = {(data["p_value"] < alpha).sum() / data.shape[0]}\n'
          f'При увеличении количества симуляций, это значение стремиться к ',
          f"заданой мощности {power}" if test == "AB" else f"заданому уровню значимости {alpha}")
```
